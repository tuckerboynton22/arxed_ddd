```python
codes = pd.read_excel('/Users/tuckerboynton/Desktop/R/ArxEd/enrollment.xlsx')['district_code'].unique()
districts = pd.read_excel('/Users/tuckerboynton/Desktop/R/ArxEd/enrollment.xlsx')['district_name'].unique()
```

```python
loop = 0
for code, district in zip(codes, districts):
    loop += 1
    if code // 10000 < 10:
        zeros = "000"
    elif code // 10000 < 100:
        zeros = "00"
    elif code // 10000 < 1000:
        zeros = "0"
    else:
        zeros = ""
    new_url = base_url + zeros + str(code) + "&orgtypecode=5"
    new_page = requests.get(new_url)
    new_page_html = new_page.text
    darts = re.findall(r'_ct[^>]*>([^*]*)', new_page_html)
    if darts == []:
        print(darts)
    print(loop)
    
    for dart in darts:
        results_df.at[district, dart] = 1
    
    sleep(30)
```
